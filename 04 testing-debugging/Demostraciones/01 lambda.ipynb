La sintaxis general es:

```python
lambda argumentos: expresion
```
Una función lambda puede recibir uno o varios argumentos, pero únicamente puede contener una expresión cuyo resultado será devuelto automáticamente.

````python
def calcular_total(cantidad, precio_unitario):
    return cantidad * precio_unitario

total = calcular_total(5, 120)

print(total)
````

````python
calcular_total = lambda cantidad, precio_unitario: cantidad * precio_unitario

total = calcular_total(5, 120)

print(total)
````

````python
pedidos = [
    {"cliente": "Contoso", "total": 1200},
    {"cliente": "Northwind", "total": 800},
    {"cliente": "Adventure Works", "total": 2500}
]

pedidos
````

````python
def obtener_total(pedido):
    return pedido["total"]

pedidos_ordenados = sorted(
    pedidos,
    key=obtener_total
)

for pedido in pedidos_ordenados:
    print(pedido)
````

````python
pedidos_ordenados = sorted(
    pedidos,
    key=lambda pedido: pedido["total"]
)

for pedido in pedidos_ordenados:
    print(pedido)
````
