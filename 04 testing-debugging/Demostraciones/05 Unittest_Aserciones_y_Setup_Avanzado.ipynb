
# 4.3.1 Métodos de Aserción y Métodos Adicionales de Configuración

En esta sección conoceremos algunos conceptos importantes del framework **unittest**.

A diferencia de las secciones anteriores, este cuaderno tiene un enfoque conceptual. Su objetivo es comprender qué herramientas existen dentro de `unittest`, cuándo se utilizan y cuál es su propósito general.

No utilizaremos código en esta lección. La intención es construir una visión de alto nivel que sirva como referencia para futuras pruebas unitarias.



# ¿Qué es una aserción?

Una aserción es una verificación automática.

Cuando escribimos una prueba unitaria, normalmente tenemos una expectativa sobre el comportamiento del sistema.

Por ejemplo:

- Esperamos que un cálculo produzca un resultado específico.
- Esperamos que una validación sea verdadera.
- Esperamos que un objeto exista.
- Esperamos que se genere un error cuando se proporciona información incorrecta.

Las aserciones permiten expresar esas expectativas de forma clara y automática.

Si la expectativa se cumple, la prueba pasa.

Si la expectativa no se cumple, la prueba falla.



# ¿Por qué existen diferentes métodos de aserción?

Aunque podríamos verificar todo utilizando comparaciones generales, `unittest` proporciona métodos especializados para expresar de forma más clara qué estamos comprobando.

Esto tiene varias ventajas:

- Las pruebas son más fáciles de leer.
- Los mensajes de error son más descriptivos.
- La intención del desarrollador queda más clara.
- El mantenimiento de las pruebas resulta más sencillo.

Cada método de aserción está diseñado para un tipo específico de comprobación.



# Principales métodos de aserción

A continuación se muestran los métodos más utilizados dentro de `unittest.TestCase`.

## assertEqual()

Verifica que dos valores sean iguales.

Es el método de aserción más utilizado porque muchas pruebas consisten en comprobar que el resultado obtenido coincide exactamente con el resultado esperado.

---

## assertNotEqual()

Verifica que dos valores sean diferentes.

Se utiliza cuando queremos asegurarnos de que un resultado no coincida con un valor específico.

---

## assertTrue()

Verifica que una condición sea verdadera.

Es útil cuando el resultado esperado es simplemente una condición lógica positiva.

---

## assertFalse()

Verifica que una condición sea falsa.

Es la contraparte de `assertTrue()`.

---

## assertIs()

Verifica que dos referencias apunten exactamente al mismo objeto.

No se enfoca en el contenido, sino en la identidad del objeto.

---

## assertIsNot()

Verifica que dos referencias no correspondan al mismo objeto.

---

## assertIsNone()

Verifica que un valor sea nulo (`None`).

Resulta muy útil cuando un método puede no devolver información.

---

## assertIsNotNone()

Verifica que exista un valor válido.

Se utiliza frecuentemente para comprobar que una búsqueda o consulta produjo resultados.

---

## assertIn()

Verifica que un elemento exista dentro de una colección o texto.

Puede utilizarse con listas, tuplas, conjuntos, diccionarios o cadenas.

---

## assertNotIn()

Verifica que un elemento no se encuentre dentro de una colección.

---

## assertIsInstance()

Verifica que un objeto pertenezca a una clase determinada.

Es especialmente útil cuando trabajamos con herencia y polimorfismo.

---

## assertNotIsInstance()

Verifica que un objeto no pertenezca a una clase específica.

---

## assertRaises()

Verifica que una operación genere una excepción determinada.

Este método es muy importante porque una buena prueba no solamente valida escenarios exitosos; también valida comportamientos incorrectos y errores esperados.



# ¿Cuál es el método de aserción más importante?

No existe un único método más importante.

Cada uno responde a una necesidad diferente.

Sin embargo, en proyectos reales suelen utilizarse con mayor frecuencia:

1. assertEqual()
2. assertTrue()
3. assertFalse()
4. assertRaises()
5. assertIsInstance()

Estos cinco métodos cubren una gran parte de las pruebas unitarias habituales.



# Métodos adicionales de configuración

En la lección anterior conocimos:

- setUp()
- tearDown()

Estos métodos se ejecutan antes y después de cada prueba individual.

Sin embargo, existen otros métodos que permiten preparar recursos en diferentes niveles.

Esto resulta especialmente útil cuando las pruebas comparten información o recursos costosos de crear.



# setUpClass() y tearDownClass()

Estos métodos trabajan a nivel de clase de pruebas.

## setUpClass()

Se ejecuta una sola vez antes de que comience la primera prueba de la clase.

Su propósito es preparar recursos compartidos que serán utilizados por todas las pruebas.

Ejemplos:

- Abrir una conexión a una base de datos.
- Cargar un archivo grande.
- Crear un conjunto de datos común.
- Iniciar un servicio temporal.

---

## tearDownClass()

Se ejecuta una sola vez cuando todas las pruebas de la clase han terminado.

Su objetivo es liberar los recursos creados por `setUpClass()`.

Ejemplos:

- Cerrar conexiones.
- Eliminar información temporal.
- Detener servicios.
- Liberar memoria ocupada por grandes estructuras de datos.



# Diferencia entre setUp() y setUpClass()

La diferencia principal está en la frecuencia de ejecución.

## setUp()

Se ejecuta antes de cada prueba.

Si existen diez pruebas, se ejecutará diez veces.

---

## setUpClass()

Se ejecuta una sola vez para toda la clase.

Si existen diez pruebas, seguirá ejecutándose solamente una vez.

Por esta razón, `setUpClass()` suele utilizarse para recursos costosos de crear y que pueden compartirse entre varias pruebas.



# setUpModule() y tearDownModule()

Existe un nivel todavía más amplio que la clase.

Ese nivel es el módulo completo.

Un módulo normalmente corresponde a un archivo de pruebas.

## setUpModule()

Se ejecuta una única vez cuando comienza la ejecución del módulo.

Permite preparar recursos que serán utilizados por múltiples clases de prueba.

---

## tearDownModule()

Se ejecuta una única vez cuando todo el módulo ha terminado.

Su función es limpiar los recursos creados al inicio.

Estos métodos suelen utilizarse en escenarios grandes donde varias clases de prueba necesitan compartir configuraciones comunes.



# Comparación de niveles de configuración

Podemos visualizar los distintos niveles de preparación de la siguiente manera:

## Nivel módulo

- setUpModule()
- tearDownModule()

Afectan a todo el archivo de pruebas.

---

## Nivel clase

- setUpClass()
- tearDownClass()

Afectan únicamente a una clase de pruebas.

---

## Nivel prueba individual

- setUp()
- tearDown()

Afectan únicamente a una prueba específica.

Esta jerarquía permite controlar exactamente cuándo se crean y destruyen los recursos necesarios para las pruebas.



# ¿Cuándo utilizar cada método?

## Usar setUp() y tearDown()

Cuando cada prueba necesita comenzar desde un estado completamente limpio.

Ejemplos:

- Crear objetos nuevos.
- Limpiar archivos temporales.
- Reiniciar información de prueba.

---

## Usar setUpClass() y tearDownClass()

Cuando varias pruebas comparten recursos costosos.

Ejemplos:

- Conexiones SQL.
- Carga de grandes conjuntos de datos.
- Configuración de servicios compartidos.

---

## Usar setUpModule() y tearDownModule()

Cuando múltiples clases necesitan utilizar la misma infraestructura.

Ejemplos:

- Configuración general de pruebas.
- Servicios comunes.
- Recursos compartidos entre varios grupos de pruebas.



# Resumen general

El framework `unittest` ofrece mucho más que `assertEqual()`.

Las aserciones permiten expresar diferentes tipos de expectativas sobre el comportamiento del sistema.

Además, los métodos de configuración permiten preparar y limpiar recursos en distintos niveles:

| Nivel | Métodos |
|---------|---------|
| Módulo | setUpModule(), tearDownModule() |
| Clase | setUpClass(), tearDownClass() |
| Prueba individual | setUp(), tearDown() |

Comprender esta estructura ayuda a construir pruebas más organizadas, eficientes y fáciles de mantener.



# Conclusión

A medida que los proyectos crecen, también crece la necesidad de automatizar pruebas y administrar correctamente los recursos utilizados durante su ejecución.

Las aserciones permiten verificar comportamientos.

Los métodos de configuración permiten preparar el entorno adecuado para realizar esas verificaciones.

Frase para recordar:

> Una buena prueba no solo valida resultados; también administra correctamente el entorno donde esos resultados son evaluados.
