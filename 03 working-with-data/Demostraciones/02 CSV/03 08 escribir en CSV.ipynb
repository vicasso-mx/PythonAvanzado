Supongamos que el área de soporte desea generar un archivo CSV con los tickets atendidos durante el día para compartirlo con otra aplicación.

```python
import csv

tickets = [
    ["Ticket", "Asunto", "Estado"],
    ["TK-001", "Error de acceso", "Atendido"],
    ["TK-002", "Impresora sin conexión", "Pendiente"],
    ["TK-003", "Restablecer contraseña", "Cerrado"]
]

with open("../data/tickets_generados.csv", "w", newline="",encoding="utf-8-sig") as archivo:

    escritor = csv.writer(archivo)

    for fila in tickets:
        escritor.writerow(fila)

print("Archivo generado.")
```

# 3.8.2. Crear un archivo CSV con `csv.DictWriter`

## ¿Qué problema resuelve?

En muchos sistemas los datos se manejan como diccionarios porque permiten acceder a los campos por nombre en lugar de por posición.

```python
tickets = [
    {
        "Ticket": "TK-004",
        "Asunto": "Error de acceso",
        "Estado": "Atendido"
    },
    {
        "Ticket": "TK-005",
        "Asunto": "Impresora sin conexión",
        "Estado": "Pendiente"
    },
    {
        "Ticket": "TK-006",
        "Asunto": "Restablecer contraseña",
        "Estado": "Cerrado"
    }
]

campos = ["Ticket", "Asunto", "Estado"]

with open(
    "../data/tickets_dict.csv",
    "w",
    newline="",
    encoding="utf-8-sig"
) as archivo:

    escritor = csv.DictWriter(
        archivo,
        fieldnames=campos
    )

    escritor.writeheader()

    escritor.writerows(tickets)

print("Archivo generado.")
```

# 3.8.3. Agregar registros a un archivo CSV (Append)

## ¿Qué problema resuelve?

Supongamos que ya existe un archivo con tickets y llega una nueva solicitud de soporte. En lugar de recrear el archivo completo, podemos agregar únicamente la nueva fila.

```python
import csv

nuevo_ticket = [
    "TK-004",
    "Configuración de correo",
    "Nuevo"
]

with open(
    "../data/tickets_generados.csv",
    "a",
    newline="",
    encoding="utf-8-sig"
) as archivo:

    escritor = csv.writer(archivo)

    escritor.writerow(nuevo_ticket)

print("Ticket agregado.")
```
