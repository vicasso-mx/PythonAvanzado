# Estructura típica de XML

Supongamos el siguiente documento XML:

```xml
<records>

    <record>
        <player alias="Baby Ruth">Babe Ruth</player>
        <homeruns>60</homeruns>
        <year>1927</year>
    </record>

    <record>
        <player  alias="MM wire">Mark McGwire</player>
        <homeruns>70</homeruns>
        <year>1998</year>
    </record>

</records>
```

# Importante: lxml convierte etiquetas a minúsculas

Si el XML contiene:

```xml
<Player>Babe Ruth</Player>
```

Beautiful Soup con `lxml` lo interpreta como:

```xml
<player>Babe Ruth</player>
```

Por ello siempre conviene buscar:

```python
soup.find_all("player")
```

y no:

```python
soup.find_all("Player")
```


````python
import requests

url = "https://static.webucator.com/media/public/documents/hrleaders.xml"

respuesta = requests.get(url)

contenido = respuesta.text
````

````python
from bs4 import BeautifulSoup

soup = BeautifulSoup(contenido, "lxml")
````

````python
registros = soup.find_all("record")

print("Total de registros:", len(registros))
````

````python
primer_registro = registros[2]

print(primer_registro)
````

````python
jugadores = soup.find_all("player")

print("Jugadores encontrados:", len(jugadores))
````

````python
for jugador in jugadores[:5]:
    print(jugador.text)
````

````python
primer_registro = registros[0]

jugador = primer_registro.find("player").text
anio = primer_registro.find("year").text

print(jugador)
print(anio)
````
