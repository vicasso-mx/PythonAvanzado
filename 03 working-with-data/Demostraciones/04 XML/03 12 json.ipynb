            }
        }

````python
import requests

# URL de ejemplo
url = "https://api.ejemplo.com/weather"

respuesta = requests.get(url)

# Convertir JSON a dict
weather = respuesta.json()

# Obtener información
ciudad = weather["city"]["name"]

temp_max = weather["list"][0]["main"]["temp_max"]

print("Ciudad:", ciudad)
print("Temperatura máxima:", temp_max)
````
