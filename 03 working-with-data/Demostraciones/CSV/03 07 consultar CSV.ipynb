# fieldnames
Los objetos csv.DictReader tienen una propiedad fieldnames que contiene una lista con las claves obtenidas de la primera fila de datos. 

````python
lineas_curso_csv = "../data/LineasCurso.csv"

with open(lineas_curso_csv, newline='', encoding='utf-8-sig') as archivo:
    lector = csv.DictReader(archivo)
    print(" | ".join(lector.fieldnames))

    for fila in lector:
            print( fila
            )
````

````python
encabezados = ['ClaveLineaCurso','ClaveLineaCursoAlterna','NombreLineaCurso']

with open(lineas_curso_csv, newline='', encoding='utf-8') as archivo:
    lector = csv.DictReader(archivo, encabezados)
    
    print(" | ".join(lector.fieldnames))
    
    for fila in lector:
        print(
            f"{fila['ClaveLineaCurso']} | "
            f"{fila['ClaveLineaCursoAlterna']} | "
            f"{fila['NombreLineaCurso']}"
        )
````

````python
import csv
from datetime import date

archivo_csv = "tickets.csv"

# Convierte "yyyy-mm-dd" a un objeto date
texto_a_fecha = lambda texto: date(*map(int, texto.split("-")))

def buscar_ticket(archivo_csv, ticket_buscado, fecha_buscada):
    with open(archivo_csv, newline="", encoding="utf-8-sig") as archivo:
        tickets = csv.DictReader(archivo)
        return next(
            (
                fila
                for fila in tickets
                if fila["Ticket"] == ticket_buscado
                and texto_a_fecha(
                    fila["FechaSolicitud"]
                ) == fecha_buscada
            ),
            None
        )

def main():

    ticket = input("Número de ticket (por ejemplo 'DL-0013250'): ").strip()

    fecha_texto = input("Fecha de solicitud (yyyy-mm-dd): ").strip()

    fecha = texto_a_fecha(fecha_texto)

    resultado = buscar_ticket(archivo_csv, ticket, fecha)

    if resultado:
        print("\nTicket encontrado:")
        print(
            f"Ticket: {resultado['Ticket']} | "
            f"Asunto: {resultado['Asunto']} | "
            f"Estado: {resultado['EstadoAtencion']}"
        )
    else:
        print(
            "\nNo se encontró el ticket."
        )

main()
````
