```python
usuarios = [
    "Ana",
    "Luis",
    "Pedro"
]

proyectos = [
    "CRM",
    "Portal",
    "BI"
]

print("Usuarios:", usuarios)
print("Proyectos:", proyectos)
```

## Permutaciones tradicionales

```python
asignaciones = []

for usuario in usuarios:
    for proyecto in proyectos:
        asignaciones.append(
            (usuario, proyecto)
        )

for asignacion in asignaciones:
    print(asignacion)
```

## Con comprensiones de lista

```python
asignaciones = [
    (usuario, proyecto)
    for usuario in usuarios
    for proyecto in proyectos
]

for asignacion in asignaciones:
    print(asignacion)
```

# Combinaciones

```python
desarrolladores = [
    "Ana",
    "Luis",
    "Pedro",
    "Sofía"
]

desarrolladores
```

# Develop tradicional

```python
parejas = []

for i in range(len(desarrolladores)):
    for j in range(i, len(desarrolladores)):
        parejas.append(
            (
                desarrolladores[i],
                desarrolladores[j]
            )
        )

for pareja in parejas:
    print(pareja)
```

```python
parejas = [
    (
        desarrolladores[i],
        desarrolladores[j]
    )
    for i in range(len(desarrolladores))
    for j in range(i, len(desarrolladores))
]

for pareja in parejas:
    print(pareja)
```
