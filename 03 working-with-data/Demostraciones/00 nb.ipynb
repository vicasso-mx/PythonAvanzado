# 3.4. SQLite
SQLite es un motor de base de datos SQL sin servidor. Cada base de datos SQLite se almacena en un único archivo que puede transportarse fácilmente entre computadoras. Aunque SQLite no es tan robusto como los sistemas empresariales de gestión de bases de datos relacionales, funciona muy bien para bases de datos locales o bases de datos que no tienen cargas de trabajo grandes.


# 3.5. Base de Datos SQLite en Memoria
Python te permite crear bases de datos en memoria con SQLite. 

Para crear una conexión a una base de datos en memoria, utiliza el siguiente código:
 
```python
connection = sqlite3.connect(':memory:')
```
