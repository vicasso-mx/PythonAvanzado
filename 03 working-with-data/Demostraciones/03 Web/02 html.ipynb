# 3.9.3. Beautiful Soup

Beautiful Soup permite analizar contenido HTML o XML para encontrar información específica dentro de una página web. Su uso común es extraer datos estructurados desde tablas, listas, enlaces o bloques de contenido. La documentación oficial describe `find_all()` como un método que devuelve todas las etiquetas o cadenas que coinciden con los filtros indicados.

```python
import requests
from bs4 import BeautifulSoup
```

`requests` obtiene el contenido de una página web y `BeautifulSoup` permite analizar ese contenido como si fuera una estructura navegable.


```python
url = "https://static.webucator.com/media/public/documents/hrleaders.html"

respuesta = requests.get(url)
contenido = respuesta.text
```

Aquí se descarga el contenido de la página. La variable `contenido` guarda el HTML como texto.

```python
print(contenido)
```

```python
soup = BeautifulSoup(contenido, "lxml")
```

El objeto `soup` representa el documento HTML ya interpretado. Desde aquí se pueden buscar etiquetas como `tr`, `td`, `th`, etc.

```python
filas = soup.find_all("tr")

print("Total de filas encontradas:", len(filas))
```

```python
filas = soup("tr")
celdas = soup("td")

print("Filas:", len(filas))
print("Celdas:", len(celdas))
```

```python
primera_fila = filas[0]

encabezados = primera_fila("th")
celdas_datos = primera_fila("td")

print("Encabezados:", encabezados)
print("Celdas de datos:", celdas_datos)
```

```python
fila_steroids = soup.find("tr", "steroids-era")

print(fila_steroids)
```

```python
ruths = soup.find_all("td", string="Babe Ruth")

print("Veces que aparece Babe Ruth:", len(ruths))
```

```python
primera_fila_ruth = ruths[0].parent

print(primera_fila_ruth)
```

Se obtienen todas las celdas de la fila y se toma la última con `[-1]`, que contiene la fecha de nacimiento.

```python
celdas_ruth = primera_fila_ruth.find_all("td")
fecha_nacimiento = celdas_ruth[-1].text

print("Fecha de nacimiento:", fecha_nacimiento)
```
