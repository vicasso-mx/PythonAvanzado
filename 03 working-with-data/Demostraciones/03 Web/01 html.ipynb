# 3.9.1. El Paquete Requests
En Python se puede instalar un paquete llamado `Requests` para realizar solicitudes HTTP, es muy amigable para los desarrolladores. 

Aunque se pueden utilizar todos los métodos de solicitud HTTP (por ejemplo, post, put, head, etc.), en la mayoría de los casos utilizarás el método get mediante requests.get(), al cual le pasarás la URL como una cadena de texto.

````python
import requests

url = 'https://static.webucator.com/media/public/documents/hrleaders.html'
r = requests.get(url)
content = r.text
print(content[:125]) # print first 125 characters
````

````python
import requests

url = 'https://static.webucator.com/media/public/documents/hrleaders.html'
r = requests.get(url, headers={'user-agent': 'my-app/0.0.1'})
content = r.text
print(content[:225]) 
````

# 3.9.2. HTML
Los datos en las páginas web se almacenan en HTML, un lenguaje de marcado relativamente simple. Los elementos de una página HTML se marcan mediante etiquetas (tags). Por ejemplo, un párrafo comienza con una etiqueta de apertura `<p>` y termina con una etiqueta de cierre `</p>`. La página mostrada presenta registros de jonrones en una tabla. Cada registro se encuentra en una fila de tabla marcada de la siguiente manera:

```html
<tr class="steroids-era">
    <td>1</td>
    <td>Mark McGwire</td>
    <td>70</td>
    <td>1998</td>
    <td>1963-10-01</td>
</tr>
<a url="1963-10-01" />
```
- Las filas de una tabla comienzan con una etiqueta de apertura `<tr>` y terminan con una etiqueta de cierre `</tr>`.
- Dentro de las filas de una tabla, las celdas de datos comienzan con una etiqueta de apertura `<td>` y terminan con una etiqueta de cierre `</td>` 
- Las celdas de encabezado comienzan con una etiqueta de apertura `<th>` y terminan con una etiqueta de cierre `</th>`.
- Los atributos normalmente vienen en pares nombre-valor.
- los atributos solo aparecen en la etiqueta de apertura, de esta forma:

```html
<tagname att1="value" att2="value">Contenido del elemento</tagname>
```

Algunas etiquetas tienen un atributo class":
```html
<tr class="steroids-era">
```

Es común buscar elementos por nombre de etiqueta y por clase.
