# Otros Atributos

Hasta ahora hemos buscado elementos por:

* Nombre de etiqueta (`tr`, `td`, `a`, etc.).
* Clase CSS (`class`).

Sin embargo, también es posible buscar elementos por cualquier atributo HTML, como:

* `href`
* `target`
* `id`
* `name`
* `src`
* `alt`

Por ejemplo:

```html
<a href="https://www.ejemplo.com" target="_blank">
    Visitar sitio
</a>
```

El atributo que nos interesa es:

```html
target="_blank"
```

````python
import requests
from bs4 import BeautifulSoup


url = "https://static.webucator.com/media/public/documents/hrleaders.html"

respuesta = requests.get(url)
soup = BeautifulSoup(respuesta.text, "lxml")

enlaces = soup.find_all("a", target="_blank")

print("Enlaces encontrados:", len(enlaces))
````

# Encontrar Imágenes Sin Atributo ALT

## ¿Por qué es importante?

Una imagen accesible debería tener un atributo `alt`.

Ejemplo correcto:

```html
<img src="logo.png" alt="Logo de la empresa">
```

Ejemplo incorrecto:

```html
<img src="logo.png">
```

Las personas que utilizan lectores de pantalla dependen del atributo `alt` para conocer el contenido o propósito de una imagen.

````python
imagenes_sin_alt = soup.find_all("img", alt=False)

print("Imágenes sin atributo alt:", len(imagenes_sin_alt))
````
