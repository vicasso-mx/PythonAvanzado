## 5. Encapsulación
La encapsulación consiste en **proteger los datos internos de un objeto** para evitar que sean modificados de forma incorrecta desde fuera de la clase.

En lugar de permitir esto:

```python
ticket.estado = "lo que sea"
```

Buscamos controlar los cambios mediante métodos o propiedades.


````python
class Ticket:
    def __init__(self, titulo, descripcion):
        self.titulo = titulo
        self.descripcion = descripcion
        self._estado = "Abierto"

    def cerrar(self):
        self._estado = "Cerrado"

    def __str__(self):
        return f"{self.titulo} - Estado: {self._estado}"
````

### 6. `@property`

`@property` permite acceder a un método como si fuera un atributo.

Sirve para exponer información interna de forma controlada.

````python
class Ticket:
    def __init__(self, titulo, descripcion):
        self.titulo = titulo
        self.descripcion = descripcion
        self._estado = "Abierto"

    @property
    def estado(self):
        return self._estado

    def cerrar(self):
        self._estado = "Cerrado"

    def __str__(self):
        return f"{self.titulo} - Estado: {self.estado}"
    
def main():
    ticket = Ticket("Error en login", "El usuario no puede iniciar sesión")

    print(ticket)

    ticket.cerrar()

    print(ticket.estado)


main()
````

````python
class Ticket:
    ESTADOS_VALIDOS = ["Abierto", "En proceso", "Cerrado"]

    def __init__(self, titulo, descripcion):
        self.titulo = titulo
        self.descripcion = descripcion
        self._estado = "Abierto"

    @property
    def estado(self):
        return self._estado

    @estado.setter
    def estado(self, nuevo_estado):
        if nuevo_estado not in self.ESTADOS_VALIDOS:
            raise ValueError("Estado no válido")
        self._estado = nuevo_estado
````

````python
def main():
    ticket = Ticket("Error en login", "El usuario no puede iniciar sesión")

    print(ticket.estado)
    
    ticket.estado = "En proceso"

    print(ticket.estado)

    ticket.estado = "Cancelado"

main()
````

### 7. Decoradores básicos

Un decorador modifica o amplía el comportamiento de una función o método sin cambiar directamente su código interno.

`@property` es un decorador.

Este código:

```python
@property
def estado(self):
    return self._estado
```

equivale conceptualmente a decir:

```python
estado = property(estado)
```

````python
def decorador(funcion):
    def envoltura():
        print("Antes de ejecutar la función")
        funcion()
        print("Después de ejecutar la función")
    return envoltura


@decorador
def saludar():
    print("Hola")
````

````python
saludar()
````

En esta fase no necesitamos dominar todos los decoradores todavía.

Lo importante es entender:

```python
@property
```

no es magia. Es una forma elegante de transformar un método para que se comporte como atributo.

````python
class Ticket:
    """
    Representa un ticket de soporte.

    Esta clase permite crear tickets con título, descripción y estado.
    También demuestra el uso de constructor, atributos, métodos,
    encapsulación, propiedades, decoradores y docstrings.
    """

    ESTADOS_VALIDOS = ["Abierto", "En proceso", "Cerrado"]

    def __init__(self, titulo, descripcion):
        """
        Constructor de la clase Ticket.

        Parámetros:
        titulo: título breve del ticket.
        descripcion: explicación del problema o solicitud.
        """
        self._titulo = titulo
        self._descripcion = descripcion
        self._estado = "Abierto"
    
    @property
    def titulo(self):
        return self._titulo

    @titulo.setter
    def titulo(self, nuevo_titulo):
        self._titulo = nuevo_titulo

    @property
    def descripcion(self):
        return self._descripcion

    @descripcion.setter
    def descripcion(self, nueva_descripcion):
        self._descripcion = nueva_descripcion

    @property
    def estado(self):
        """
        Devuelve el estado actual del ticket.

        Esta propiedad permite consultar el estado sin acceder
        directamente al atributo interno _estado.
        """
        return self._estado

    @estado.setter
    def estado(self, nuevo_estado):
        """
        Modifica el estado del ticket solo si es válido.
        """
        if nuevo_estado not in self.ESTADOS_VALIDOS:
            raise ValueError(f"Estado no válido: {nuevo_estado}")

        self._estado = nuevo_estado

    def cambiar_estado(self, nuevo_estado):
        """
        Cambia el estado del ticket usando la propiedad estado.
        """
        self.estado = nuevo_estado

    def cerrar(self):
        """
        Cambia el estado del ticket a Cerrado.
        """
        self.estado = "Cerrado"

    def __str__(self):
        """
        Devuelve una representación legible del ticket.
        """
        return f"Ticket: {self.titulo} | Estado: {self.estado}"


def main():
    """
    Función principal del programa.

    Aquí se crean y utilizan las instancias de la clase Ticket.
    """

    # Crear instancias
    ticket_1 = Ticket(
        "Error en inicio de sesión",
        "El usuario no puede entrar al sistema."
    )

    ticket_2 = Ticket(
        "Reporte no genera PDF",
        "El botón de exportar no responde."
    )

    # Mostrar tickets usando __str__
    print(ticket_1)
    print(ticket_2)

    # Consultar propiedad
    print("Estado inicial:", ticket_1.estado)

    # Cambiar estado usando método
    ticket_1.cambiar_estado("En proceso")
    print("Estado actualizado:", ticket_1.estado)

    # Cerrar ticket
    ticket_1.cerrar()
    print("Estado final:", ticket_1.estado)

    # Mostrar ayuda de la clase
    help(Ticket)


if __name__ == "__main__":
    main()
````
