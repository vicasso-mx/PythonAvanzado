# Tema 1. Crear un Ticket (Clase y Objeto)

## Situación de negocio

**¿Cómo opera una mesa de ayuda (Help Desk)?**. Cuando un usuario reporta un problema, se crea un ticket. Por ejemplo:

```text
TK-1001
No puedo acceder al correo
Alta
Abierto
```

Cada ticket posee información propia.

## ¿Qué es una clase?

Una clase es un molde. Gracias a una clase podemos definir que las características y acciones de una entidad, sin tener aún una identidad propia.

````python
class Ticket:
    pass
````

````python
ticket1 = Ticket() # los paréntesis es el constructor de clase
ticket2 = Ticket()
````

````python
class Ticket:
    # declaraciones
    pass

ticket1 = Ticket()

print(ticket1)
````

````python
class Ticket:
    
    # definiendo un constructor __init__ con la siguente firma: (self, folio, asunto, prioridad): 
    def __init__(self, folio, asunto, prioridad): # constructor
        self.folio = folio
        self.asunto = asunto
        self.prioridad = prioridad
        self.estado = "Abierto"
````

````python
ticket = Ticket("TK-1001", "No puedo acceder al correo", "Alta")
````

````python
class Ticket:

    def __init__(self, folio, asunto, prioridad):
        self.folio = folio
        self.asunto = asunto
        self.prioridad = prioridad
        self.estado = "Abierto"

    def cerrar(self):
        # lógica más compleja, reglas, flujo, autorizaciones, bifurcaciones, etc.
        self.estado = "Cerrado"
````

````python
ticket = Ticket(
    "TK-1001",
    "No puedo acceder al correo",
    "Alta"
)

print(ticket.estado)
ticket.cerrar()
print(ticket.estado)
````

````python
print(ticket)
````

No es útil para un usuario.

Python permite personalizar la representación textual mediante:

```python
__str__()
```

````python
class Ticket:

    def __init__(self, folio, asunto, prioridad):
        self.folio = folio
        self.asunto = asunto
        self.prioridad = prioridad
        self.estado = "Abierto"

    def __str__(self):
        return (
            f"{self.folio} | "
            f"{self.asunto} | "
            f"{self.estado}"
        )
````

````python
ticket = Ticket(
    "TK-1001",
    "No puedo acceder al correo",
    "Alta"
)

print(ticket)
````

# Tema 4. Docstrings

## ¿Qué es un Docstring?

Es una descripción integrada dentro de una clase o método.

Permite documentar el código profesionalmente.

````python
class Ticket:
    """
    Representa un ticket de soporte técnico.

    Permite almacenar información del incidente
    y controlar su estado.
    """

    def __init__(self, folio, asunto, prioridad):
        """
        Inicializa un nuevo ticket.

        Args:
            folio (str): Identificador único.
            asunto (str): Descripción breve.
            prioridad (str): Nivel de prioridad.
        """

        self.folio = folio
        self.asunto = asunto
        self.prioridad = prioridad
        self.estado = "Abierto"
````

````python
help(Ticket)
````

````python
print(Ticket.__doc__)
````
