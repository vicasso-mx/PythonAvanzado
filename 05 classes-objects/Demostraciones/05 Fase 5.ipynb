# Fase 5 — Reutilización mediante Herencia
### 15. Herencia

La herencia permite crear una nueva clase a partir de otra.

La nueva clase hereda de la clase padre:

* atributos
* métodos
* propiedades
* comportamiento

## Aplicado a tickets

Tenemos una clase general:

```python
class TicketBase:
    pass
```

Y varias especializaciones:

```python
class TicketSoporte(TicketBase):
    pass
```

```python
class TicketDesarrollo(TicketBase):
    pass
```

```python
class TicketInfraestructura(TicketBase):
    pass
```


significa:

> hereda de TicketBase

````python
class TicketAbstracto:
    pass
````

````python
class TicketBase(TicketAbstracto):

    def __init__(self, titulo):
        self.titulo = titulo
        self.estado = "Abierto"

    def cerrar(self):
        self.estado = "Cerrado"
````

````python
class TicketSoporte(TicketBase):
    pass


ticket = TicketSoporte("No puedo iniciar sesión")

print(ticket.titulo)

ticket.cerrar()

print(ticket.estado)
````

````python
class TicketBase:
    def atender(self):
        print("Atendiendo ticket")

class TicketSoporte(TicketBase):
    def cerrar(self): 
        pass

class TicketDesarrollo(TicketBase):
    def autorizar(self): 
        pass
````

````python
t1 = TicketSoporte()
t2 = TicketDesarrollo()

t1.atender
t2.atender
````

````python
class TicketBase:
    def atender(self):
        print("Atendiendo ticket")

class TicketSoporte(TicketBase):
    def atender(self):
        print("Atendiendo incidente de soporte")

class TicketDesarrollo(TicketBase):
    def atender(self):
        print("Analizando requerimiento de desarrollo")
````

````python
t1 = TicketSoporte()
t2 = TicketDesarrollo()

t1.atender
t2.atender
````

# 17. super()

Ahora aparece una situación interesante.

Queremos reutilizar parte del comportamiento del padre.

````python
# sin super
class TicketSoporte(TicketBase):

    def __init__(self, titulo):
        self.titulo = titulo
        self.estado = "Abierto"
        self.prioridad = "Alta"
````

````python
# super
class TicketSoporte(TicketBase):

    def __init__(self, titulo):
        super().__init__(titulo)
        self.prioridad = "Alta"
````

## ¿Qué hace super()?

Llama al padre.

Es equivalente a decir:

> "Ejecuta primero la lógica heredada."
