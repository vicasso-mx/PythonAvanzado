# Aplicado a tickets

Queremos registrar cuándo se ejecuta un método importante, por ejemplo:

```python
ticket.atender()
```

Podemos crear un decorador llamado:

```python
registrar_accion
```


````python
def registrar_accion(funcion):
    def envoltura(*args, **kwargs):
        print(f"Ejecutando método: {funcion.__name__}")

        resultado = funcion(*args, **kwargs)

        print(f"Método finalizado: {funcion.__name__}")

        return resultado

    return envoltura
````

````python
class TicketSoporte:

    @registrar_accion
    def atender(self):
        return f"Ticket de soporte atendido."
    

t1 = TicketSoporte()

t1.atender()
````

````python
from functools import wraps
````

````python
def registrar_accion(funcion):
    @wraps(funcion)
    def envoltura(*args, **kwargs):
        print(f"[LOG] Ejecutando: {funcion.__name__}")

        resultado = funcion(*args, **kwargs)

        print(f"[LOG] Finalizado: {funcion.__name__}")

        return resultado

    return envoltura
````

````python
@registrar_accion
def cerrar(self):
    self.estado = "Cerrado"

@registrar_accion
def cambiar_estado(self, nuevo_estado):
    self.estado = nuevo_estado
````
