# 8. Detectar patrones comunes

## Problema

Muchos principiantes crean clases porque sí.

Por ejemplo:

```python
class Nombre:
    pass

class Telefono:
    pass

class Correo:
    pass
```

Técnicamente es válido.

Pero probablemente sea un mal diseño.

## Regla práctica

Cuando varios datos siempre viajan juntos:

```text
Nombre
Correo
Teléfono
```

probablemente pertenecen a una misma entidad.

Ejemplo:

```python
class Cliente:

    def __init__(self, nombre, correo, telefono):
        self.nombre = nombre
        self.correo = correo
        self.telefono = telefono
```

En analisis y diseño:

identificamos:
- sustantivos: se convierten en clase o no
- verbos: se convierten en metodos
- sustantivos que califican (adjetivos calificativos): atributos


## En nuestro sistema de tickets

Tenemos:

```text
Ticket
Título
Descripción
Estado
```


Todos pertenecen naturalmente a:

```python
class Ticket
```


## Primer patrón importante

Pregúntate:

> ¿Esto tiene identidad propia?

Por ejemplo:

```text
Cliente
Ticket
Empleado
Producto
Factura
Proyecto
```

Sí tienen identidad.

Normalmente terminan siendo clases.

## Segundo patrón importante

Pregúntate:

> ¿Esto es solamente un dato?

Ejemplos:

```text
Nombre
Correo
Precio
Cantidad
Fecha
```


No suelen ser clases inicialmente.

Normalmente son atributos.

## Ejemplo

Mal diseño:

```python
class Precio:
    pass

class Nombre:
    pass

class Estado:
    pass

class Ticket:
    pass
```


Mejor:

```python
class Ticket:

    def __init__(self, titulo, descripcion):
        self.titulo = titulo
        self.descripcion = descripcion
        self.estado = "Abierto"
```

# 9. ¿Qué debería existir primero?


## Error común

Pensar primero en las pantallas.

Por ejemplo:

```text
Pantalla de Tickets
Pantalla de Usuarios
Pantalla de Reportes
```

Y después intentar descubrir las clases.

## En Clases y Objetos hacemos lo contrario

Primero identificamos los objetos del negocio.

Luego construimos el sistema alrededor de ellos.


## Caso Ticket

Imagina que el cliente dice:

> Necesitamos registrar solicitudes de soporte.

¿Qué aparece inmediatamente?

```text
Solicitud
Ticket
Usuario
```

````python
class Ticket:
    pass

class Usuario:
    pass
````

Todavía vacías.

Pero ya existen.

````python
class Ticket:

    def cerrar(self):
        pass
````

````python
class Ticket:

    def __init__(self, titulo, descripcion):
        self.titulo = titulo
        self.descripcion = descripcion
````

````python
@property
def estado(self):
    pass
````

## Evolución correcta

```text
Problema de negocio
        ↓
Entidades
        ↓
Clases
        ↓
Atributos
        ↓
Métodos
        ↓
Reglas
        ↓
Optimización
```
