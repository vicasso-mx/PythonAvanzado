# 1. Inicio y Fin (^ $)

## ¿Qué problema resuelve?

Permite validar que un valor completo tenga un formato específico y no solamente que contenga una palabra.

debe comenzar con TK

````python
import re

ejemplos = [
    "TK-123",
    "TKABC",
    "ABC-TK",
    "XXTK",
    "TK",
    "ticket",
    "TK-456",
]

cumplen = [valor for valor in ejemplos if re.search(r"^TK", valor)]
no_cumplen = [valor for valor in ejemplos if not re.search(r"^TK", valor)]

print("EJEMPLO 1 - ^TK")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
print()
````

$ ayuda a validar que debe terminar con ".pdf"

````python
archivos = [
    "manual.pdf",
    "reporte.xlsx",
    "contrato.pdf",
    "pdf_manual",
    "archivo.pdf.bak"
]

cumplen = [valor for valor in archivos if re.search(r"\.pdf$", valor)]
no_cumplen = [valor for valor in archivos if not re.search(r"\.pdf$", valor)]

print("EJEMPLO 2 - .pdf$")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
print()
````

Debe ser exactamente TK-123

````python
ejemplos = [
    "TK-123",
    "TK-1234",
    "ABC TK-123",
    "TK-123 XYZ",
    "TK-123 "
]

cumplen = [valor for valor in ejemplos if re.search(r"^TK-123$", valor)]
no_cumplen = [valor for valor in ejemplos if not re.search(r"^TK-123$", valor)]

print("EJEMPLO 3 - ^TK-123$")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
print()
````

# 2. Límites de Palabra (\b y \B)

Permiten identificar palabras completas evitando coincidencias parciales.

Ejemplo empresarial: detectar el estado ERROR sin confundirlo con ERROR123.

````python
import re

texto = "ERROR ERROR123"

print(re.findall(r"ERROR\b", texto))
````

# 3. Cantidad de Ocurrencias (?, +, *, {n})

Estos operadores controlan cuántas veces puede repetirse un patrón.

````python
import re

ejemplos = [ "TK-", "TK-1",  "TK-12", "TK-123", "TK-1234", "TK1", "TK12", "TK123", "TK1234"]
````

````python
tickets_guion_opcional = [
    valor for valor in ejemplos 
    if re.match(r"^TK-\d{3}$", valor)
    ]
print(tickets_guion_opcional)
````

````python
import re

tickets = [
    valor for valor in ejemplos
    if re.match(r"^TK-?\d{3}$", valor)
]

print(tickets)
````

````python
resultado = [
    valor for valor in ejemplos
    if re.match(r"^TK-\d+$", valor)
]

print(resultado)
````

````python
resultado = [
    valor for valor in ejemplos
    if re.match(r"^TK-\d*$", valor)
]

print(resultado)
````

# 4. Caracteres Comunes (. \d \w \s)

Son los bloques básicos más utilizados en proyectos reales.

`\d` busca dígitos, `\w` busca caracteres alfanuméricos y `\s` busca espacios en blanco

````python
import re

ejemplos = [
    "Factura123",
    "Pedido",
    "ABC999",
    "Cliente",
    "2025"
]

cumplen = [texto for texto in ejemplos if re.search(r"\d", texto)]
no_cumplen = [texto for texto in ejemplos if not re.search(r"\d", texto)]

print("Contienen al menos un dígito")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

````python
import re

ejemplos = [
    "ABC123",
    "@@@",
    "cliente_01",
    "***",
    "Python"
]

cumplen = [texto for texto in ejemplos if re.search(r"\w", texto)]
no_cumplen = [texto for texto in ejemplos if not re.search(r"\w", texto)]

print("Contienen caracteres alfanuméricos")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

````python
import re

ejemplos = [
    "Juan Perez",
    "Maria",
    "Soporte Tecnico",
    " Python",
    "Base de Datos"
]

cumplen = [texto for texto in ejemplos if re.search(r"\s", texto)]
no_cumplen = [texto for texto in ejemplos if not re.search(r"\s", texto)]

print("Contienen espacios")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

````python
import re

ejemplos = [
    "A",
    "123",
    "@",
    "",
    "Python"
]

cumplen = [texto for texto in ejemplos if re.search(r".", texto)]
no_cumplen = [texto for texto in ejemplos if not re.search(r".", texto)]

print("Contienen al menos un carácter")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

Validación de código de empleado usando \w

Debe contener exactamente:

3 letras
guion bajo
3 números

Ejemplos válidos: ABC_123, XYZ_999

````python
import re

empleados = [
    "ABC_123",
    "XYZ_999",
    "AB_123",
    "ABC123",
    "ABC_12",
    "abc_456"
]

cumplen = [
    codigo
    for codigo in empleados
    if re.match(r"^\w{3}_\d{3}$", codigo)
]

no_cumplen = [
    codigo
    for codigo in empleados
    if not re.match(r"^\w{3}_\d{3}$", codigo)
]

print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

# 5. Clases de Caracteres (`[]`)

Las **clases de caracteres** permiten definir un conjunto de caracteres válidos en una posición específica.

Son muy utilizadas para validar:

* Códigos.
* Identificadores.
* Placas.
* Folios.
* Usuarios.
* Contraseñas.

 `[ABC]`

Busca una sola letra que sea **A**, **B** o **C**.

````python
import re

ejemplos = [
    "A",
    "B",
    "C",
    "D",
    "Z"
]

cumplen = [valor for valor in ejemplos if re.search(r"[ABC]", valor)]
no_cumplen = [valor for valor in ejemplos if not re.search(r"[ABC]", valor)]

print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

````python

import re

ejemplos = [
    "A",
    "M",
    "Z",
    "a",
    "1",
    "@"
]

cumplen = [valor for valor in ejemplos if re.search(r"[A-Z]", valor)]
no_cumplen = [valor for valor in ejemplos if not re.search(r"[A-Z]", valor)]

print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

Validar código de producto

Formato:

* 3 letras mayúsculas
* guion
* 4 números

Ejemplos válidos:

* `ABC-1234`
* `XYZ-9999`

````python
import re

productos = [
    "ABC-1234",
    "XYZ-9999",
    "AB-1234",
    "abc-1234",
    "ABC1234",
    "AAA-999"
]

cumplen = [
    codigo
    for codigo in productos
    if re.match(r"^[A-Z]{3}-[0-9]{4}$", codigo)
]

no_cumplen = [
    codigo
    for codigo in productos
    if not re.match(r"^[A-Z]{3}-[0-9]{4}$", codigo)
]

print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

## 6. Negación (`[^]`)

Permite indicar caracteres que **NO deben aparecer** en una posición determinada.

En este ejemplo, un sistema de soporte técnico recibe identificadores de ticket. El formato esperado es:

* Un carácter inicial que **no sea `#`**
* Seguido de uno o más dígitos

Patrón:

```regex
[^#]\d+
```

````python
import re

tickets = [
    "T123",
    "A456",
    "#789",
    "X1",
    "#1000",
    "B999"
]

cumplen = [
    ticket
    for ticket in tickets
    if re.fullmatch(r"[^#]\d+", ticket)
]

no_cumplen = [
    ticket
    for ticket in tickets
    if not re.fullmatch(r"[^#]\d+", ticket)
]

print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
````

## 7. Grupos (`()`)

Los grupos permiten **capturar partes específicas de un texto** para poder utilizarlas posteriormente, analizarlas o extraer información de manera estructurada.

En este ejemplo, un sistema de ventas almacena números de orden con el formato:

```text
ORD-2025-1500
```

Donde:

* `ORD` = prefijo de la orden.
* `2025` = año.
* `1500` = consecutivo.

Patrón:

```regex
(ORD)-(\d{4})-(\d+)
```

````python
import re

orden = "ORD-2025-1500"

resultado = re.search(r"(ORD)-(\d{4})-(\d+)", orden)

if resultado:
    print("Texto completo:", resultado.group(0))
    print("Prefijo:", resultado.group(1))
    print("Año:", resultado.group(2))
    print("Consecutivo:", resultado.group(3))
````

## 8. Alternativas (`|`)

Las alternativas permiten indicar que una expresión puede coincidir con **una opción u otra**.

En este ejemplo, un sistema de soporte técnico clasifica tickets según su prioridad. Solo se consideran válidos los tickets que comiencen con:

* `ALTA`
* `MEDIA`
* `BAJA`

Patrón:

```regex
^(ALTA|MEDIA|BAJA)-\d+$
```

````python
import re

tickets = [
    "ALTA-1001",
    "MEDIA-2050",
    "BAJA-999",
    "URGENTE-500",
    "CRITICA-100",
    "ALTA1001"
]

cumplen = [
    ticket
    for ticket in tickets
    if re.fullmatch(r"(ALTA|MEDIA|BAJA)-\d+", ticket)
]

no_cumplen = [
    ticket
    for ticket in tickets
    if not re.fullmatch(r"(ALTA|MEDIA|BAJA)-\d+", ticket)
]

print("Cumplen:")
print(cumplen)

print("\nNo cumplen:")
print(no_cumplen)
````

## 9. Escape (`\`)

En este ejemplo, un sistema almacena documentos PDF. Queremos identificar únicamente archivos que terminen exactamente en `.pdf`.

Patrón:

```regex
\.pdf$
```


Patrón:

```regex
\.pdf$
```

Patrón:

```regex
^\$\d+
```


````python
archivos = [
    "manual.pdf",
    "reporte.xlsx",
    "contrato.pdf",
    "pdf_manual",
    "archivo.pdf.bak"
]

cumplen = [valor for valor in archivos if re.search(r".pdf$", valor)]
no_cumplen = [valor for valor in archivos if not re.search(r".pdf$", valor)]

print("EJEMPLO 2 - .pdf$")
print("Cumplen:", cumplen)
print("No cumplen:", no_cumplen)
print()
````
