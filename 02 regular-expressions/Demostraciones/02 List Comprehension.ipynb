La sintaxis básica es:

```python
mi_lista = [f(x) for x in iterable if condicion]
```

En esta estructura:

- `x` representa cada elemento que se recorre.
- `iterable` puede ser una lista, tupla, cadena, diccionario u otra colección.
- `condicion` permite filtrar elementos.
- `f(x)` representa lo que queremos devolver por cada elemento válido.

`f(x)` puede ser:

- Una variable, por ejemplo: `x`.
- Una operación, por ejemplo: `x ** 2`.
- Una llamada a función, por ejemplo: `len(x)` o `calcular_total(x)`.

````python
oportunidades = [
    {"cliente": "Contoso", "estado": "Ganada", "monto": 15000},
    {"cliente": "Northwind", "estado": "Perdida", "monto": 8000},
    {"cliente": "Adventure Works", "estado": "Ganada", "monto": 22000},
    {"cliente": "Fabrikam", "estado": "En proceso", "monto": 12000},
    {"cliente": "Tailspin Toys", "estado": "Ganada", "monto": 9500}
]

oportunidades
````

obtener únicamente aquellas que están **ganadas** para después generar un reporte ejecutivo.


## Código tradicional

````python
oportunidades_ganadas = []

for oportunidad in oportunidades:
    if oportunidad["estado"] == "Ganada":
        oportunidades_ganadas.append(oportunidad)

for oportunidad in oportunidades_ganadas:
    print(oportunidad)
````

````python
oportunidades_ganadas = [
    oportunidad
    for oportunidad in oportunidades
    if oportunidad["estado"] == "Ganada"
]

for oportunidad in oportunidades_ganadas:
    print(oportunidad)
````

en lugar de devolver todo el diccionario de cada oportunidad, podemos devolver únicamente el nombre del cliente y el monto.

## Código tradicional

````python
resumen_ganadas = []

for oportunidad in oportunidades:
    if oportunidad["estado"] == "Ganada":
        resumen = {
            "cliente": oportunidad["cliente"],
            "monto": oportunidad["monto"]
        }
        resumen_ganadas.append(resumen)

for resumen in resumen_ganadas:
    print(resumen)
````

## Lo mismo pero comprensiones de lista

````python
resumen_ganadas = [
    {
        "cliente": oportunidad["cliente"],
        "monto": oportunidad["monto"]
    }
    for oportunidad in oportunidades
    if oportunidad["estado"] == "Ganada"
]

for resumen in resumen_ganadas:
    print(resumen)
````

También podemos llamar una función dentro de una comprensión de listas.

Esto es útil cuando queremos mantener la comprensión corta y mover parte de la lógica a una función con nombre claro.


````python
def formatear_oportunidad(oportunidad):
    return f'{oportunidad["cliente"]}: ${oportunidad["monto"]:,.2f}'


reporte_ganadas = [
    formatear_oportunidad(oportunidad)
    for oportunidad in oportunidades
    if oportunidad["estado"] == "Ganada"
]

for linea in reporte_ganadas:
    print(linea)
````
