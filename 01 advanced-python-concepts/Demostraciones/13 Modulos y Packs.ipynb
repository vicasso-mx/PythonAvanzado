# 1.11. Modules and Packages


## Caso de negocio

Imaginemos un sistema comercial dividido en:

- ventas.py
- clientes.py
- reportes.py
- utilidades.py

Cada archivo tiene una responsabilidad específica.


Cada módulo contiene una variable especial llamada `__name__`.


## ¿Qué problema resuelve?

Permite distinguir si un archivo:

- Fue ejecutado directamente.
- Fue importado desde otro módulo.

````python
def prueba():
    print("Ejecutando pruebas")

if __name__ == "__main__":
    prueba()
````

# 1.11.2 Packages

## ¿Qué es un paquete?

Un paquete es una carpeta que contiene módulos relacionados.

Tradicionalmente incluye un archivo llamado:

```text
__init__.py
```


## ¿Qué hace __init__.py?

Puede:

- Estar vacío.
- Inicializar el paquete.
- Definir elementos públicos.
