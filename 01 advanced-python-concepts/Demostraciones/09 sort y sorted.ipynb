# 1.7. Sorting
- Usar `sort()` para ordenar listas en sitio.
- Usar `sorted()` para obtener una nueva lista ordenada.

# ¿Qué problema resuelve?

En un sistema real, los datos rara vez llegan en el orden que necesitamos.

Por ejemplo, una lista de ventas puede venir en el orden en que se capturó, pero para un reporte quizá necesitamos verla:

- De mayor a menor importe.
- Por nombre de cliente.
- Por vendedor.
- Por margen.
- Por estado.
- Por fecha de cierre.

El ordenamiento permite transformar datos desorganizados en información más fácil de analizar.

# 1.7.1. Sorting Lists in Place

## ¿Qué significa ordenar en sitio?

Ordenar una lista en sitio significa modificar directamente la lista original.

En Python, esto se hace con el método `sort()`.

Cuando usamos `sort()`, no se crea una nueva lista; la lista existente cambia su orden.

## Caso de negocio: ordenar productos por nombre

Supongamos que tenemos una lista de productos capturados en diferente orden.

Queremos ordenarlos alfabéticamente para mostrarlos en un reporte.

```python
productos = ["Monitor", "Laptop", "Teclado", "Mouse", "Impresora"]

productos.sort()

print(productos)
```

## Caso de negocio: ordenar ventas de mayor a menor

Supongamos que queremos identificar rápidamente las ventas más altas del mes.

```python
ventas = [8500, 12500, 4000, 22000, 9800, 15700]

ventas.sort(reverse=True)

print(ventas)
```

## Caso de negocio: ordenar productos por longitud del nombre

Queremos ordenar los productos según la longitud de su nombre.

Esto puede ser útil, por ejemplo, para ajustar visualizaciones, etiquetas o reportes donde los nombres largos requieren atención especial.


```python
productos = ["Monitor", "Laptop", "Teclado", "Mouse", "Impresora multifuncional"]

def obtener_longitud(nombre_producto):
    return len(nombre_producto)

productos.sort(key=obtener_longitud)

print(productos)
```

### Usando funciones `lambda` con `key`

```python
productos = ["Monitor", "Laptop", "Teclado", "Mouse", "Impresora multifuncional"]

productos.sort(key=lambda nombre_producto: len(nombre_producto) )

print(productos)
```

# Combinando `key` y `reverse`


```python
productos = ["Monitor", "Laptop", "Teclado", "Mouse", "Impresora multifuncional"]

productos.sort(key=lambda nombre_producto: len(nombre_producto), reverse=True )

print(productos)
```

## Caso de negocio: ordenar ventas por importe

Tenemos una lista de ventas y queremos ordenarlas de menor a mayor importe.

```python
ventas = [
    {"folio": "V-001", "cliente": "Comercial ABC", "importe": 18500}, # 0
    {"folio": "V-003", "cliente": "Servicios Norte", "importe": 22500}, # 1
    {"folio": "V-002", "cliente": "Industrias Delta", "importe": 7400}, # 2
    {"folio": "V-004", "cliente": "Grupo Solaris", "importe": 9800}, # 3
]

ventas.sort(key=lambda venta: venta[1], reverse=False)

for venta in ventas:
    print(venta["folio"], venta["cliente"], venta["importe"])
```

# 1.7.2. The `sorted()` Function

## ¿Qué problema resuelve?

El método `sort()` modifica la lista original.

Esto puede ser un problema si necesitamos conservar el orden original de los datos.

La función `sorted()` resuelve este caso porque devuelve una nueva lista ordenada sin modificar la colección original.

## Caso de negocio: conservar el orden original de captura

Supongamos que una lista de ventas está en el orden en que fueron registradas.

Queremos generar una vista ordenada por importe, pero sin perder el orden original.

```python
ventas = [8500, 12500, 4000, 22000, 9800, 15700]

ventas_ordenadas = sorted(ventas)

print("Ventas originales:", ventas)
print("Ventas ordenadas:", ventas_ordenadas)
```

```python
ventas = [8500, 12500, 4000, 22000, 9800, 15700]

ventas_ordenadas = sorted(ventas, reverse=True)

print("Ventas originales:", ventas)
print("Ventas ordenadas:", ventas_ordenadas)
```

```python
productos = ["Monitor", "Laptop", "Teclado", "Mouse", "Impresora multifuncional"]

productos_por_longitud = sorted(productos, key=lambda producto: len(producto))

print("Productos originales:", productos)
print("Productos por longitud:", productos_por_longitud)
```
