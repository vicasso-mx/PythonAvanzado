# 1.6. Mutable and Immutable Built-in Objects

En esta sección estudiaremos una diferencia fundamental en Python:

- Objetos **mutables**
- Objetos **inmutables**

A primera vista, la diferencia parece sencilla:

- Un objeto mutable puede modificarse.
- Un objeto inmutable no puede modificarse.

Sin embargo, comprender este tema con más profundidad ayuda a evitar errores comunes en proyectos reales, especialmente cuando trabajamos con listas, diccionarios, parámetros de funciones y estructuras de datos compartidas.

# ¿Qué problema resuelve este conocimiento?

En proyectos reales, muchas veces compartimos datos entre funciones, módulos o procesos.

Por ejemplo:

- Una lista de productos.
- Un diccionario con datos de cliente.
- Una configuración del sistema.
- Una colección de ventas.
- Una lista de usuarios autorizados.

Si no entendemos cuándo Python modifica un objeto existente y cuándo crea un objeto nuevo, podemos provocar errores difíciles de detectar.

Un caso muy común es modificar una lista pensando que estamos trabajando con una copia, cuando en realidad dos variables apuntan al mismo objeto.

# Objetos inmutables

Los objetos inmutables no pueden modificarse en el mismo lugar de memoria.

Algunos ejemplos comunes son:

- `str`
- `int`
- `float`
- `bool`
- `tuple`

Cuando parece que estamos modificando un objeto inmutable, Python en realidad crea un nuevo objeto.

## Ejemplo con cadenas de texto

Supongamos que tenemos el nombre de una categoría comercial.

Queremos convertirla a mayúsculas para mostrarla en un reporte.


```python
categoria = "laptops"

print(categoria)
print(id(categoria))

categoria = categoria.upper()

print(categoria)
print(id(categoria))
```

## Análisis del ejemplo

Aunque usamos la misma variable `categoria`, el objeto original no se modificó.

Lo que ocurrió fue lo siguiente:

1. Primero, `categoria` apuntaba al texto `"laptops"`.
2. Después, `categoria.upper()` creó una nueva cadena `"LAPTOPS"`.
3. La variable `categoria` comenzó a apuntar a ese nuevo objeto.

Por eso los valores de `id()` son diferentes.

Esto confirma que las cadenas de texto son inmutables.
