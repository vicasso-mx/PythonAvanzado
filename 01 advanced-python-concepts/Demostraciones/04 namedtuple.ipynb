# Comencemos

Una `namedtuple` es una tupla cuyos elementos tienen nombre.


# Tupla tradicional

```python
producto = ("Laptop Lenovo", 18500, 12)

print(producto[0])
print(producto[1])
print(producto[2])
```

## Tupla nombrada

```python
from collections import namedtuple

Producto = namedtuple("Producto", ["nombre", "precio", "existencia"])

producto = Producto("Laptop Lenovo", 18500, 12)

print(producto.nombre)
print(producto.precio)
print(producto.existencia)
```

Se puede acceder a los valores por posición o nombre

```python
from collections import namedtuple

Producto = namedtuple("Producto", "nombre precio existencia")

producto = Producto("Monitor 24 pulgadas", 4200, 8)

print("Acceso por posición:")
print(producto[0])
print(producto[1])
print(producto[2])

print("\nAcceso por nombre:")
print(producto.nombre)
print(producto.precio)
print(producto.existencia)
```

Ahora usemos una `namedtuple` para representar ventas de una empresa.

Cada venta tendrá:

1. `folio`
1. `cliente`
1. `producto`
1. `cantidad`
1. `precio_unitario`

```python
from collections import namedtuple

Venta = namedtuple(
    "Venta",
    ["folio", "cliente", "producto", "cantidad", "precio_unitario"]
)

venta = Venta(
    folio=1001,
    cliente="Comercial ABC",
    producto="Laptop Lenovo",
    cantidad=3,
    precio_unitario=18500
)

print(f"Folio: {venta.folio}")
print(f"Cliente: {venta.cliente}")
print(f"Producto: {venta.producto}")
print(f"Cantidad: {venta.cantidad}")
print(f"Precio unitario: ${venta.precio_unitario:,.2f}")
```

## Calcular información derivada

Una `namedtuple` no solo sirve para imprimir datos. También podemos utilizar sus valores en cálculos.

Por ejemplo, podemos calcular el total de una venta multiplicando la cantidad por el precio unitario.

```python
total = venta.cantidad * venta.precio_unitario

print(f"Total de la venta: ${total:,.2f}")
```

## Lista de ventas usando namedtuple

```python
from collections import namedtuple

Venta = namedtuple(
    "Venta",
    ["folio", "cliente", "producto", "cantidad", "precio_unitario"]
)

ventas = [
    Venta(1001, "Comercial ABC", "Laptop Lenovo", 3, 18500),
    Venta(1002, "Servicios Delta", "Monitor 24 pulgadas", 5, 4200),
    Venta(1003, "Grupo Norte", "Teclado inalámbrico", 10, 650),
]

for venta in ventas:
    total = venta.cantidad * venta.precio_unitario
    print(f"Folio: {venta.folio} | Cliente: {venta.cliente} | Total: ${total:,.2f}")
```

# Si uso un diccionario

```python
venta_diccionario = {
    "folio": 1001,
    "cliente": "Comercial ABC",
    "producto": "Laptop Lenovo",
    "cantidad": 3,
    "precio_unitario": 18500
}

print(venta_diccionario["cliente"])
print(venta_diccionario["precio_unitario"])
```

### Validad la inmutabilidad de una tupla nombrada

```python
from collections import namedtuple

Producto = namedtuple("Producto", ["nombre", "precio", "existencia"])
producto = Producto("Laptop Lenovo", 18500, 12)

# Esta línea provocaría un error porque namedtuple es inmutable:
#producto.precio = 17000

print(producto)
```

### Usar _replace()

```python
from collections import namedtuple

Producto = namedtuple("Producto", ["nombre", "precio", "existencia"])

producto_original = Producto("Laptop Lenovo", 18500, 12)
producto_actualizado = producto_original._replace(precio=17500) # esto devuelve un objeto nuevo.

print("Producto original:")
print(producto_original)

print("\nProducto actualizado:")
print(producto_actualizado)
```

También podemos convertir una `namedtuple` en un diccionario utilizando el método `_asdict()`.


```python
from collections import namedtuple

Venta = namedtuple("Venta", ["folio", "cliente", "total"])

venta = Venta(1001, "Comercial ABC", 55500)

venta_como_diccionario = venta._asdict()

print(venta_como_diccionario)
```

## ¿Cuándo conviene usar namedtuple?

Conviene usar `namedtuple` cuando:

- Los datos tienen una estructura fija.
- No necesitas modificar los valores directamente.
- **Quieres algo más claro que una tupla tradicional.**
- Buscas una alternativa más ligera que una clase.
- Estás representando registros simples de negocio.

Ejemplos adecuados:

- Una venta procesada.
- Un producto consultado de un catálogo.
- Un empleado leído desde un reporte.
- Una fila de resultados obtenida de una consulta.
- Un registro temporal usado para cálculos.

## ¿Cuándo no conviene usar namedtuple?

No suele ser la mejor opción cuando:

- Los datos cambian constantemente.
- Necesitas agregar comportamiento complejo.
- Requieres validaciones internas avanzadas.
- El objeto necesita muchos métodos propios.
- Los datos tienen muchas variaciones de estructura.

En esos casos puede ser mejor utilizar:

- Diccionarios.
- Clases.
- `dataclasses`.
- Modelos especializados según el framework utilizado.
