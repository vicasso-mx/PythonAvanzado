# 1.8. Sorting Sequences of Sequences


## Caso de negocio: ventas como tuplas

Supongamos que cada venta se representa como una tupla:

```python
(cliente, importe)
```

Primero veremos cómo Python ordena automáticamente estas tuplas.

````python
ventas = [
    ("Servicios Norte", 22500),
    ("Comercial ABC", 18500),
    ("Grupo Solaris", 9800),
    ("Industrias Delta", 7400),
]

ventas_ordenadas = sorted(ventas)

print(ventas_ordenadas)
````

````python
ventas = [
    ("Servicios Norte", 22500),
    ("Comercial ABC", 18500),
    ("Grupo Solaris", 9800),
    ("Industrias Delta", 7400),
]

ventas_por_importe = sorted(ventas, key=lambda venta: venta[1])

print(ventas_por_importe)
````

````python
ventas = [
    ("Norte", "Servicios Norte", 22500),
    ("Centro", "Comercial ABC", 18500),
    ("Norte", "Grupo Solaris", 9800),
    ("Centro", "Industrias Delta", 7400),
]

ventas = sorted(ventas, key=lambda venta: (venta[0], venta[2]))

print(ventas)
````

````python
clientes = ["grupo solaris", "Comercial ABC", "industrias delta", "Servicios Norte"]

clientes_ordenados = sorted(clientes)

print(clientes_ordenados)
````

````python
clientes = ["grupo solaris", "Comercial ABC", "industrias delta", "Servicios Norte"]

clientes_ordenados = sorted(clientes, key=lambda cliente: cliente.lower())

print(clientes_ordenados)
````

## Caso de negocio: ordenar tickets de soporte

Supongamos que tenemos tickets de soporte con prioridad y tiempo de resolución.

Queremos ordenar los tickets por prioridad y después por tiempo de resolución.

````python
tickets = [
    {"id": 101, "cliente": "Comercial ABC", "prioridad": 2, "tiempo_horas": 5},
    {"id": 102, "cliente": "Industrias Delta", "prioridad": 1, "tiempo_horas": 12},
    {"id": 103, "cliente": "Servicios Norte", "prioridad": 3, "tiempo_horas": 3},
    {"id": 104, "cliente": "Grupo Solaris", "prioridad": 1, "tiempo_horas": 9},
]

tickets_ordenados = sorted(
    tickets,
    key=lambda ticket: (ticket["prioridad"], ticket["tiempo_horas"])
)

for ticket in tickets_ordenados:
    print(ticket["id"], ticket["cliente"], ticket["prioridad"], ticket["tiempo_horas"])
````
