# 1.5. Mapping and Filtering

En esta sección trabajaremos con dos funciones integradas de Python:

- `map()`
- `filter()`

Ambas permiten procesar colecciones de datos de forma secuencial, pero con propósitos diferentes:

- `map()` se usa para **transformar** datos.
- `filter()` se usa para **seleccionar** datos que cumplen una condición.

Trabajaremos con un escenario de negocio relacionado con ventas, productos e indicadores comerciales.

# 1.5.1. `map(function, iterable, ...)`

## ¿Qué problema resuelve?

En proyectos reales es común necesitar transformar todos los elementos de una colección.

Por ejemplo:

- Convertir precios sin IVA a precios con IVA.
- Normalizar nombres de productos.
- Calcular comisiones de vendedores.
- Transformar importes de una moneda a otra.
- Generar etiquetas o descripciones a partir de datos existentes.

La función `map()` permite aplicar una función a cada elemento de uno o más iterables y devuelve un iterador con los resultados.

En términos simples:

> `map()` toma una colección, aplica una transformación a cada elemento y produce una nueva secuencia transformada.

## Caso de negocio

Supongamos que una empresa tiene una lista de precios sin IVA y necesita calcular los precios finales con IVA incluido.

Para simplificar el ejemplo, usaremos una tasa de IVA del 16%.

### Código tradicional

```python
precios_sin_iva = [1200, 850, 4300, 2199, 15000]

precios_con_iva = []

for precio in precios_sin_iva:
    precio_final = precio * 1.16
    precios_con_iva.append(precio_final)

print(precios_con_iva)
```

### Misma solución con .map()

```python
def calcular_precio_con_iva(precio):
    return precio * 1.16


precios_sin_iva = [1200, 850, 4300, 2199, 15000]

resultado = map(calcular_precio_con_iva, precios_sin_iva)

print(resultado)
```

## Importante: `map()` devuelve un iterador

El resultado de `map()` no es directamente una lista.

Python devuelve un objeto iterable que puede recorrerse, pero si queremos ver todos los valores de inmediato, normalmente lo convertimos con `list()`.

```python
def calcular_precio_con_iva(precio):
    return precio * 1.16

precios_sin_iva = [1200, 850, 4300, 2199, 15000]

resultado = list(map(calcular_precio_con_iva, precios_sin_iva))

print(resultado)
```

## Ventajas de `map()`

`map()` puede ser útil cuando:

- Existe una transformación clara que se aplicará a todos los elementos.
- Ya tienes una función definida que realiza esa transformación.
- Deseas expresar que todos los elementos serán procesados bajo la misma regla.
- Quieres trabajar con iteradores en lugar de crear listas inmediatamente.

Sin embargo, no siempre es obligatorio usar `map()`. En muchos casos, una list comprehension puede ser igual o más clara.

## Alternativa con list comprehension

Python permite resolver el mismo problema con una list comprehension.

Esta alternativa suele ser muy legible cuando la operación es corta.

```python
precios_sin_iva = [1200, 850, 4300, 2199, 15000]

precios_con_iva = [precio * 1.16 for precio in precios_sin_iva]

print(precios_con_iva)
```

## `map()` con múltiples iterables

`map()` también puede trabajar con más de un iterable.

Esto es útil cuando una función necesita recibir valores de dos o más listas relacionadas.

Por ejemplo, supongamos que tenemos:

- Una lista de precios.
- Una lista de descuentos aplicables.

Queremos calcular el precio final de cada producto después del descuento.

```python
def aplicar_descuento(precio, descuento):
    return precio * (1 - descuento)

precios = [1000, 2500, 800, 12000]
descuentos = [0.10, 0.15, 0.05, 0.20]

precios_finales = list(map(aplicar_descuento, precios, descuentos))

print(precios_finales)
```

## Precaución con múltiples iterables

Cuando `map()` recibe varios iterables, se detiene al llegar al final del iterable más corto.

Esto puede ser conveniente en algunos casos, pero también puede ocultar errores si esperabas que ambas listas tuvieran la misma longitud.

```python
def aplicar_descuento(precio, descuento):
    return precio * (1 - descuento)

precios = [1000, 2500, 800, 12000]
descuentos = [0.10, 0.15]

precios_finales = list(map(aplicar_descuento, precios, descuentos))

print(precios_finales)
```

```python
precios = [1000, 2500, 800, 12000]
descuentos = [0.10, 0.15]

if len(precios) != len(descuentos):
    print("Error: las listas de precios y descuentos no tienen la misma longitud.")
else:
    precios_finales = list(map(aplicar_descuento, precios, descuentos))
    print(precios_finales)
```

# 1.5.2. `filter(function, iterable)`

## ¿Qué problema resuelve?

Mientras `map()` transforma elementos, `filter()` selecciona elementos.

En proyectos reales es muy común necesitar quedarnos únicamente con los registros que cumplen cierta condición.

Por ejemplo:

- Productos con inventario bajo.
- Ventas mayores a cierto importe.
- Clientes activos.
- Facturas vencidas.
- Tickets de soporte abiertos.
- Empleados con cumplimiento menor al objetivo.

## Caso de negocio

Supongamos que tenemos una lista de ventas y queremos obtener únicamente las ventas mayores o iguales a $10,000.

Primero resolveremos el problema con un ciclo `for`.
Después lo resolveremos con `filter()`.

## Forma tradicional

```python
ventas = [8500, 12500, 4000, 22000, 9800, 15700]

ventas_importantes = []

for venta in ventas:
    if venta >= 10000:
        ventas_importantes.append(venta)

print(ventas_importantes)
```

## Solución utilizando `filter()`

Para usar `filter()`, primero definimos una función que devuelva `True` o `False`.

En este caso, la función indicará si una venta es importante.

```python
def es_venta_importante(venta):
    return venta >= 10000

ventas = [8500, 12500, 4000, 22000, 9800, 15700]

ventas_importantes = list(filter(es_venta_importante, ventas))

print(ventas_importantes)
```

## `lambda` con `map()`

Supongamos que queremos calcular precios con IVA, pero no necesitamos reutilizar la función en otra parte.

Podemos escribir la transformación directamente dentro de `map()`.



```python
def calcular_precio_con_iva(precio):
    return precio * 1.16

precios_sin_iva = [1200, 850, 4300, 2199, 15000]

resultado = list(map(calcular_precio_con_iva, precios_sin_iva))

print(resultado)
```

```python
precios_sin_iva = [1200, 850, 4300, 2199, 15000]

resultado = list(map(lambda precio: precio * 1.16, precios_sin_iva))

print(resultado)
```

## `lambda` con `filter()`

Ahora filtremos ventas importantes usando una función `lambda`.

```python
def es_venta_importante(venta):
    return venta >= 10000

ventas = [8500, 12500, 4000, 22000, 9800, 15700]

ventas_importantes = list(filter(es_venta_importante, ventas))

print(ventas_importantes)
```

```python
ventas = [8500, 12500, 4000, 22000, 9800, 15700]

ventas_importantes = list(filter(lambda venta: venta >= 10000, ventas))

print(ventas_importantes)
```
