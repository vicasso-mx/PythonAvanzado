# 1.4 Counters
Uso de `Counter` para conteo de ocurrencias en escenarios de negocio.

## ¿Qué problema resuelve Counter?

Cuando necesitamos contar elementos repetidos en una colección.

```python
from collections import Counter

categorias = [
    "Laptop","Laptop","Monitor","Laptop",
    "Teclado","Monitor","Teclado"
]

conteo = Counter(categorias)
print(conteo)
```

## Actualizando contadores

El objeto `Counter` no es inmutable, es una subclase de `dict` y por lo tanto se puede modificar después de su creación.

Una de las ventajas de `Counter` es que puede acumular nuevos elementos sin necesidad de recorrer listas ni verificar si una clave existe previamente.

```python
from collections import Counter

ventas = Counter(["Laptop", "Monitor", "Laptop"])
ventas.update(["Laptop", "Teclado"])
print(ventas)
```

## Actualizando con otro Counter

En escenarios empresariales es común que la información provenga de múltiples fuentes.

Por ejemplo:

* Una sucursal registra sus ventas.
* Otra sucursal registra las suyas.
* Posteriormente ambas deben consolidarse.

```python
from collections import Counter

sucursal_a = Counter(["Laptop","Laptop","Monitor"])
sucursal_b = Counter(["Laptop","Teclado"])

sucursal_a.update(sucursal_b)
print(sucursal_a)
```

## Restando Cantidades

Además de sumar información, `Counter` permite disminuir valores mediante el método `subtract()`.

Este método resulta útil cuando necesitamos reflejar movimientos negativos como:

* Ventas que disminuyen inventario.
* Cancelaciones de pedidos.
* Devoluciones de productos.
* Ajustes de existencias.

El comportamiento es similar a `update()`, pero los valores se restan en lugar de sumarse.

```python
from collections import Counter

inventario = Counter({
    "Laptop": 10,
    "Monitor": 5
})

ventas = Counter({
    "Laptop": 3,
    "Monitor": 7
})

inventario.subtract(ventas)

print(inventario)
```

## El método most_common()

En muchos proyectos no basta con conocer los conteos; también es importante identificar cuáles son los elementos más frecuentes.

El método `most_common()` devuelve los elementos ordenados desde el más frecuente hasta el menos frecuente.

Es especialmente útil para responder preguntas como:

* ¿Cuáles son los productos más vendidos?
* ¿Qué categorías generan más movimientos?
* ¿Cuáles son los errores más comunes registrados en el sistema?
* ¿Qué clientes realizan más compras?

Si se especifica un número, el método devolverá únicamente esa cantidad de resultados.

```python
from collections import Counter

productos = [
    "Laptop","Laptop","Monitor","Laptop",
    "Teclado","Monitor","Mouse","Laptop"
]

conteo = Counter(productos)
print(conteo.most_common(3))
```
