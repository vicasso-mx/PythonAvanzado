# 1.10. Unpacking Sequences in Function Calls
Aprenderemos a utilizar el operador `*` para desempaquetar listas y tuplas al llamar funciones.

## ¿Qué problema resuelve?

En muchos sistemas los datos ya vienen agrupados en listas o tuplas, pero las funciones esperan parámetros separados. El desempaquetado permite enviar esos valores automáticamente.

```python
def calcular_total(subtotal, impuestos, envio):
    return subtotal + impuestos + envio

print(calcular_total(1000,160,120))
```

## Caso de negocio

La información de una venta proviene de un sistema externo y llega en una tupla.

```python
venta = (1000, 160, 120)
print(venta)
```

## Solución tradicional


```python
def calcular_total(subtotal, impuestos, envio):
    return subtotal + impuestos + envio

venta = (1000,160,120)

print(calcular_total(venta[0], venta[1], venta[2]))
```

El operador `*` convierte cada elemento de la secuencia en un argumento independiente.

```python
def calcular_total(subtotal, impuestos, envio):
    return subtotal + impuestos + envio

venta = (1000,160,120)

print(calcular_total(*venta))
```

```python
def generar_resumen(cliente, importe, margen):
    print(f'Cliente: {cliente}')
    print(f'Importe: ${importe:,.2f}')
    print(f'Margen: {margen:.0%}')

registro = ('Comercial ABC',18500,0.22)

generar_resumen(*registro)
```

## Precaución

La cantidad de elementos debe coincidir con la cantidad de parámetros esperados.

```python
def calcular_total(subtotal, impuestos, envio):
    return subtotal + impuestos + envio

datos = (1000,160)
calcular_total(*datos)
```

```python
def evaluar_oportunidad(cliente, importe, probabilidad):
    valor_esperado = importe * probabilidad
    print(f'{cliente}: ${valor_esperado:,.2f}')

oportunidad = ('Corporativo Lima',310000,0.50)

evaluar_oportunidad(*oportunidad)
```
