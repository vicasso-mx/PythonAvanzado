# 1.9. Creating a Dictionary from Two Sequences


Es común recibir información en listas separadas, por ejemplo productos y precios. Un diccionario permite relacionar ambas estructuras y consultar datos rápidamente.

```python
productos = ['Laptop Lenovo','Monitor 24 pulgadas','Teclado inalámbrico','Mouse ergonómico']
precios = [18500,4200,850,620]

catalogo = {}
for i in range(len(productos)):
    catalogo[productos[i]] = precios[i]

print(catalogo)
```

## Solución con zip()

`zip()` agrupa elementos de la misma posición en forma de tuplas.

```python
productos = ['Laptop Lenovo','Monitor 24 pulgadas','Teclado inalámbrico','Mouse ergonómico']
precios = [18500,4200,850,620]

print(list(zip(productos, precios)))
```

```python
productos = ['Laptop Lenovo','Monitor 24 pulgadas','Teclado inalámbrico','Mouse ergonómico']
precios = [18500,4200,850,620]

print(dict(zip(productos, precios)))
```

## Caso de negocio

Relacionar empleados con departamentos.

```python
empleados = ['Ana Torres','Luis Pérez','María Gómez','Carlos Ruiz']
departamentos = ['Finanzas','Ventas','Operaciones','TI']

empleado_departamento = dict(zip(empleados, departamentos))
print(empleado_departamento)
```

## Longitudes diferentes

```python
productos = ['Laptop','Monitor','Teclado','Mouse']
precios = [18500,4200]

print(dict(zip(productos, precios)))
```

```python
if len(productos) != len(precios):
    print('Error: las listas no tienen la misma longitud.')
```
