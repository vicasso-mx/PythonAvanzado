## 1. El problema con los diccionarios tradicionales

En Python, un diccionario permite almacenar información mediante pares de clave y valor.

Por ejemplo, podríamos almacenar el total vendido por zona:

````python
ventas_por_zona = {
    "Norte": 15000,
    "Centro": 22000
}

print(ventas_por_zona["Norte"])
````

El problema aparece cuando intentamos consultar o modificar una clave que todavía no existe.

Por ejemplo, si intentamos incrementar las ventas de la zona `Sur`, pero esa zona aún no está registrada en el diccionario, Python genera un error.

````python
ventas_por_zona = {
    "Norte": 15000,
    "Centro": 22000
}

# Esta línea genera un KeyError porque la clave "Sur" aún no existe.
ventas_por_zona["Sur"] += 5000
````

## 2. Caso de negocio: acumular ventas por zona

Supongamos que tenemos una lista de ventas realizadas por diferentes zonas comerciales.

Cada venta incluye:

- folio;
- cliente;
- zona;
- total de la venta.

Queremos calcular cuánto se vendió en total por cada zona.

````python
ventas = [
    {"folio": 1001, "cliente": "Comercial ABC", "zona": "Norte", "total": 12500},
    {"folio": 1002, "cliente": "Industrias Delta", "zona": "Centro", "total": 18700},
    {"folio": 1003, "cliente": "Servicios Roma", "zona": "Norte", "total": 9300},
    {"folio": 1004, "cliente": "Grupo Solaris", "zona": "Sur", "total": 22100},
    {"folio": 1005, "cliente": "Comercial ABC", "zona": "Centro", "total": 7600},
    {"folio": 1006, "cliente": "Logística Maya", "zona": "Sur", "total": 15400},
]

ventas
````

## 3. Solución con diccionario tradicional e `if-else`


````python
ventas_por_zona = {}

for venta in ventas:
    zona = venta["zona"]
    total = venta["total"]

    if zona in ventas_por_zona:
        ventas_por_zona[zona] += total
    else:
        ventas_por_zona[zona] = total

ventas_por_zona
````

## 4. Solución con `try-except`

````python
ventas_por_zona = {}

for venta in ventas:
    zona = venta["zona"]
    total = venta["total"]

    try:
        ventas_por_zona[zona] += total
    except KeyError:
        ventas_por_zona[zona] = total

ventas_por_zona
````

`defaultdict` es una variante de diccionario incluida en el módulo `collections`.

La diferencia principal es que, cuando consultamos una clave que no existe, `defaultdict` la crea automáticamente usando un valor predeterminado.

Para usarlo, primero debemos importarlo.

Esto significa que, cuando una clave no exista, Python la creará automáticamente con el valor que devuelve `int()` sin argumentos.

Ese valor es `0`.

````python
from collections import defaultdict

ventas_por_zona = defaultdict(int)

for venta in ventas:
    zona = venta["zona"]
    total = venta["total"]

    ventas_por_zona[zona] += total

ventas_por_zona
````

## 9. `defaultdict` con listas

Otro uso muy común de `defaultdict` consiste en agrupar elementos.

Por ejemplo, supongamos que queremos agrupar las ventas por zona, pero no solamente sumar el total, sino guardar las ventas completas de cada zona.

Para eso podemos usar `defaultdict(list)`.

````python
ventas_agrupadas_por_zona = defaultdict(list)

for venta in ventas:
    zona = venta["zona"]
    ventas_agrupadas_por_zona[zona].append(venta)

ventas_agrupadas_por_zona
````

## 10. `defaultdict` con `lambda`

También podemos usar una función `lambda` para definir un valor inicial personalizado.

Por ejemplo, supongamos que queremos construir un resumen por zona con dos datos:

- total vendido;
- número de operaciones.

Cada zona deberá iniciar con esta estructura:

```python
{"total": 0, "operaciones": 0}

````python
resumen_por_zona = defaultdict(lambda: {"total": 0, "operaciones": 0})

for venta in ventas:
    zona = venta["zona"]
    total = venta["total"]

    resumen_por_zona[zona]["total"] += total
    resumen_por_zona[zona]["operaciones"] += 1

resumen_por_zona
````

## 11. `defaultdict` con una función definida por el usuario

En lugar de usar una función `lambda`, también podemos crear una función normal.

Esto es recomendable cuando el valor inicial tiene más estructura o cuando queremos que el código sea más fácil de leer.

````python
def crear_resumen_cliente():
    return {
        "total": 0,
        "operaciones": 0,
        "folios": []
    }

resumen_por_cliente = defaultdict(crear_resumen_cliente)

for venta in ventas:
    cliente = venta["cliente"]
    total = venta["total"]
    folio = venta["folio"]

    resumen_por_cliente[cliente]["total"] += total
    resumen_por_cliente[cliente]["operaciones"] += 1
    resumen_por_cliente[cliente]["folios"].append(folio)

resumen_por_cliente
````

## 12. ¿Cuándo usar `defaultdict`?

`defaultdict` es útil cuando necesitamos:

- acumular importes;
- contar elementos;
- agrupar registros;
- construir estructuras anidadas;
- evitar validaciones repetitivas de existencia de claves.

En proyectos de negocio aparece con frecuencia al procesar información como:

- ventas por zona;
- tickets por estatus;
- empleados por departamento;
- productos por categoría;
- pagos por cliente;
- incidencias por prioridad.

## 13. ¿Cuándo NO conviene usar `defaultdict`?

Aunque es muy útil, no siempre es la mejor opción.

Puede ser preferible usar un diccionario normal cuando:

- quieres detectar explícitamente si una clave no existe;
- una clave faltante representa un error de datos;
- necesitas validar la información antes de crear nuevos registros;
- quieres evitar que una simple consulta genere claves nuevas automáticamente.
